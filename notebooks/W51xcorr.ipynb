What about the $\chi^2$ approach?$\chi^2 = \Sigma_{ij} \frac{(X_{ij}-Y_{ij})^2}{\sigma_{ij}^2}$ assuming Y is an error-free model$$ = \Sigma_{ij} \left[ X_{ij}^2/\sigma_{ij}^2 - 2X_{ij}Y_{ij}/\sigma_{ij} + Y_{ij}^2 \right]$$ $$\Sigma_{ij}[x,y] \frac{X_{ij}^2}{\sigma_{ij}^2} = X/dx * X/dx$$$$\Sigma_{ij}[x,y] \frac{X_{ij}Y_{ij}}{\sigma_{ij}} = X/dx * Y$$$$\Sigma_{ij}[x,y] Y_{ij}^2 = Y * Y$$But this isn't quite right!  X is never shifted from X and Y is never shifted from Y; these terms are the *peak* of the autocorrelation, not the autocorrelation!

```python
from AG_fft_tools import correlate2ddef make_chi2img(im1,im2, boundary='wrap'):    #ac1 = correlate2d(im1,im1, boundary=boundary)    xc = correlate2d(im1,im2, boundary=boundary)    ac1peak = (im1**2).sum()    ac2peak = (im2**2).sum()    #ac2 = correlate2d(im2,im2, boundary=boundary)    return ac1peak - 2*xc + ac2peak
```
